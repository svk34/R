# Support Vector Machine:
### It creates a hyperplane or line and divide the data into classes. 

```r
model_svm<-svm(y_b ~ . ,data=train)
summary(model_svm)
```

```r
svm_pred<-predict(model_svm,test,type = "response")
confusionMatrix(as.factor(test$y_b),as.factor(svm_pred),mode = "prec_recall")
```
